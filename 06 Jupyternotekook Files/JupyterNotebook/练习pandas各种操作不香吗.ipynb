```python
import pandas as pd 
import numpy as np 

df = pd.read_csv('job_info1.csv',engine="python",header=None)
df
```

```python
df.index = range(len(df))
df.columns = ["岗位名","公司名","工作地点","工资","发布日期","经验与学历","公司类型","公司规模","工作描述"]
```

```python
df
```

```python
print('去重之前的记录数',df.shape)
df.drop_duplicates(subset=["公司名","岗位名"],inplace=True)
print('去重之后记录数',df.shape)
```

```python
df['岗位名'].value_counts()
df['岗位名']= df['岗位名'].apply(lambda x:x.lower())
```

```python
df.shape
df = ['算法', '开发', '分析', '工程师', '数据', '运营', '运维']
index = [df["岗位名"].str.count(i) for i in df]
index = np.array(index).sum(axis=0) > 0
df = df[index]
df.shape
```

```python
ban=df['班级'].unique()
writer = pd.ExcelWriter('./多条件筛选并保存多sheet.xlsx')
for index in range(len(ban)):
    df['语文']=df.loc[df['语文'] > 70]
    df.loc[df['班级']==ban[index]].to_excel(writer,sheet_name=ban[index],index=False)
writer.save()
writer.close()
```
